## Container with Most Water

**Method #1:** Iterative Approach
- Time Complexity : `O(n)`
- Space Complexity : `O(1)`

```python
def container_with_most_water(arr: list) -> int:
    left, right = 0, len(arr) - 1
    area = 0
    
    while left < right:
        # Calculate the potential area
        temp = min(arr[left], arr[right]) * (right - left)          # REMEMBER
        # Update the maximum area if current area is greater
        if temp > area:
            area = temp
            
        # Move the pointer pointing to the shorter line
        if arr[left] <= arr[right]:
            left += 1
        else:
            right -= 1
            
    return area
```

```python
arr = [1, 8, 6, 2, 5, 4, 8, 3, 7]
```

```python
container_with_most_water(arr)
```
