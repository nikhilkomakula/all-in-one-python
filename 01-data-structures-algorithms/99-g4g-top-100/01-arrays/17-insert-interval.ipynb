## [Insert in sorted and non-overlapping interval array](https://www.geeksforgeeks.org/insert-in-sorted-and-non-overlapping-interval-array/)

#### Given a set of non-overlapping intervals and a new interval, insert the interval at the correct position. If the insertion results in overlapping intervals, then merge the overlapping intervals. Assume that the set of non-overlapping intervals is sorted based on start time, to find the correct position of insertion.

- Example 1:
    - Input: Set : [1, 3], [6, 9]
    - New Interval : [2, 5]
    - Output: [1, 5], [6, 9]
    - Explanation: The correct position to insert a new interval [2, 5] is between the two given intervals.
    - The resulting set would have been [1, 3], [2, 5], [6, 9], but the intervals [1, 3], [2, 5] are overlapping. So, they are merged in one interval [1, 5].
- Example 2:
    - Input: Set : [1, 2], [3, 5], [6, 7], [8, 10], [12, 16]
    - New Interval : [4, 9]
    - Output: [1, 2], [3, 10], [12, 16]
    - Explanation: First the interval is inserted between intervals [3, 5] and [6, 7]. Then overlapping intervals are merged together in one interval.

**Method #1:** Iterative Approach
- Time Complexity : `O(n)`
- Space Complexity : `O(n)`

```python
def insert_interval(ei: list[list[int]], ni: list[int]) -> list[list[int]]:
    i = 0
    n = len(ei)
    result = []
    
    while i < n and ei[i][0] < ni[0]:
        result.append(ei[i])
        i += 1
        
    if not result or result[-1][1] < ni[0]:
        result.append(ni)
    else:
        result[-1][1] = max(result[-1][1], ni[1])
        
    while i < n:
        if result[-1][1] < ei[i][0]:
            result.append(ei[i])
        else:
            result[-1][1] = max(result[-1][1], ei[i][1])
        i += 1
        
    return result
```

```python
ei = [[1, 3], [5, 7], [6, 12]]
ni = [2, 6]
```

```python
print(insert_interval(ei, ni))
```
