## Overlapping Intervals

**Method #1:** Iterative Approach
- Time Complexity : `O(n + m)`
- Space Complexity : `O(n + m)`

```python
def overlapping_intervals(la: list, lb: list) -> list:
    result = []
    i = j = 0
    
    while i < len(la) and j < len(lb):
        start = max(la[i][0], lb[j][0])
        end = min(la[i][1], lb[j][1])
        if start <= end:
            result.append([start, end])
        
        # compare which interval 'ends' earlier and move forward in that list
        if la[i][1] < lb[j][1]:                                                 # REMEMBER `1`
            i += 1
        else:
            j += 1
    return result
```

```python
la = [[1,4],[5,6],[7,8],[9,15]]
lb = [[2,4],[5,7],[9,15]]
```

```python
print(overlapping_intervals(la, lb))
```
