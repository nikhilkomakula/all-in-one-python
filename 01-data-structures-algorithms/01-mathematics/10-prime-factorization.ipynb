## Prime Factorization

Given a number n, write an efficient function to print all prime factors of n. For example, if the input number is 12, then output should be “2 2 3”. And if the input number is 315, then output should be “3 3 5 7”.

- Time Complexity: `O(n * sqrt(n))`
    - The time complexity of the prime_factorization function is O(n * sqrt(n)), where n is the input number. This is because the function iterates through all numbers from 2 to n in the worst case scenario, and for each number, it checks if it is prime using the is_prime function which has a time complexity of O(sqrt(n)). 
- Space Complexity: `O(1)`

```python
# checks if the given number is a prime number or not
def is_prime(n: int) -> bool:
    if n <= 1:
        return False
    elif n <= 3:
        return True
    
    if n % 2 == 0 or n % 3 == 0:
        return False
    
    i = 2
    while (i * i <= n):
        if n % i == 0:
            return False
        i += 1
    return True
```

```python
def prime_factorization(n: int) -> None:
    for i in range(2, n):
        if is_prime(i):
            temp = i
            while n % temp == 0:
                print(i)
                temp *= i
```
